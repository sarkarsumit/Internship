Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb-includes- earpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC
TSVZAXUHGREPBFGI&marketplace.
When you will open the above link you will reach to the below shown webpage .

As shown in the above page you have to scrape the tick marked attributes.These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100 reviews.
Note: All the steps required during scraping should be done through code only and not manually.

```python
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb-includes- earpods-poweradapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC TSVZAXUHGREPBFGI&marketplace.")
time.sleep(3)

# To click on all reviews
click_button = driver.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div/div/div[5]/div/a/div")
click_button.click()
time.sleep(2)

# Creating empty lists
ratings = []
review_summarys = []
full_summarys = []

# Scrapping all the info by using while loop
i = 1
while i < 11:
    rating_tags = driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']") # Ratings tag
    for j in rating_tags:
        ratings.append(j.text)
    review_summary_tags = driver.find_elements_by_xpath("//p[@class='_2-N8zT']") # Review summary tag
    for k in review_summary_tags:
        review_summarys.append(k.text)
    full_summary_tags = driver.find_elements_by_class_name("t-ZTKy") # Full summary tag
    for l in full_summary_tags:
        full_summarys.append(l.text)
    i += 1

    # creating a data frame
df = pd.DataFrame({'Ratings':ratings, 'Review Summary':review_summarys , 'Full Summary':full_summarys})
df
```
