### Q1: Write a python program to display all the header tags from wikipedia.org.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
headers = [] # Empty List
for i in soup.find_all('span', class_='mw-headline'):
    headers.append(i.text)
    
headers
```

```python
import pandas as pd
```

```python
Headers = pd.DataFrame({'Headers': headers})
Headers
```

### Q2: Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page2 = requests.get("https://www.imdb.com/list/ls009997493/")
page2
```

```python
soup2 = BeautifulSoup(page2.content)
```

```python
name = [] # Empty List
for i in soup2.find_all('h3', class_="lister-item-header"):
    name.append(i.text.split('\n')[2])
    
name
```

```python
year_of_release = [] # Empty List
for i in soup2.find_all('h3', class_="lister-item-header"):
    year_of_release.append(i.text.split('\n')[3])
    
year_of_release
```

```python
rating_test = soup2.find('span', class_="ipl-rating-star__star")
rating_test
```

```python
rating = [] # Empty List
for i in soup2.find_all('div', class_="ipl-rating-star small"):
    rating.append(i.text.replace('\n',''))
    
rating
```

```python
df1 = pd.DataFrame({'Name':name, 'Year_of_Release':year_of_release, 'Rating':rating})
df1.head(100) # Since it's already 100 rows so we can remove (.head(100))
```

### Q: Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp .

```python
page3 = requests.get("https://meesho.com/bags-ladies/pl/p7vbp")
page3
```

```python
soup3 = BeautifulSoup(page3.content)
```

```python
product_name = [] # Empty List
for i in soup3.find_all('div', class_="Card__BaseCard-sc-b3n78k-0 dUNFgg NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS"):
    product_name.append(i.text.split('₹')[0])
    
product_name
```

```python
price = [] # Empty List
for i in soup3.find_all('div', class_="Card__BaseCard-sc-b3n78k-0 dUNFgg NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS"):
    price.append(i.text.split('₹')[1])
    
price
```

```python
discount = [] # Empty List
for i in soup3.find_all('div', class_="Card__BaseCard-sc-b3n78k-0 dUNFgg NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS"):
    discount.append(i.text.split('₹')[3].split()[0])
    
discount
```

```python
df2 = pd.DataFrame({'Product_Name':product_name, 'Price':price, 'Discount':discount})
df2
```

### Q5: Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
matches = []
for i in soup4.find_all('tr', class_='table-body'):
    matches.append(i.text.split()[3])
matches
```

```python
points = []
for i in soup4.find_all('tr', class_='table-body'):
    points.append(i.text.split()[4])
points
```

```python
rating2 = []
for i in soup4.find_all('tr', class_='table-body'):
    rating2.append(i.text.split()[5])
rating2
```

```python
df3 = pd.DataFrame({'Team':team_name, 'Matches':matches, 'Points':points, 'Rating':rating2})
df3.head(10)
```

### Q7: Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.


```python
page6 = requests.get("https://coreyms.com/")
page6
```

```python
heading = []
for i in soup6.find_all('h2', class_='entry-title'):
    heading.append(i.text)
heading
```

```python
df6 = pd.DataFrame({'Heading':heading, 'Date':date, 'Content':content})
df6
```
