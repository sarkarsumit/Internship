### Q1: Write a python program to display all the header tags from wikipedia.org.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
headers = [] # Empty List
for i in soup.find_all('span', class_='mw-headline'):
    headers.append(i.text)
    
headers
```

```python
import pandas as pd
```

```python
Headers = pd.DataFrame({'Headers': headers})
Headers
```

### Q2: Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page2 = requests.get("https://www.imdb.com/search/title/?groups=top_250&sort=user_rating")
page2
```

```python
soup2 = BeautifulSoup(page2.content)
```

```python
name = [] # Empty List
for i in soup2.find_all('h3', class_="lister-item-header"):
    name.append(i.text.split('\n')[2])
    
name
```

```python
year_of_release = [] # Empty List
for i in soup2.find_all('h3', class_="lister-item-header"):
    year_of_release.append(i.text.split('\n')[3])
    
year_of_release
```

```python
rating_test = soup2.find('span', class_="global-sprite rating-star imdb-rating")
rating_test
```

```python
rating = [] # Empty List
for i in soup2.find_all('span', class_="ratings-bar"):
    rating.append(i.text)
    
rating
```
